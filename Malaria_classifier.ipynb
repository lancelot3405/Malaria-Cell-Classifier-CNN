```python
# --- 1. Imports ---
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense, Conv2D, MaxPool2D, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image

# Evaluation and Visualization Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
import seaborn as sns
import pandas as pd
import os

print(f"TensorFlow Version: {tf.__version__}")

# --- 2. Configuration and Data Paths ---
# Image dimensions (standardized input size for the CNN)
IMG_WIDTH, IMG_HEIGHT = 64, 64
BATCH_SIZE = 32
EPOCHS = 10 # Increase to 15 or 20 for potentially better results
MODEL_FILENAME = 'malaria_cnn_model.h5'

# !!! IMPORTANT: UPDATE THESE PATHS !!!
TRAIN_DATA_DIR = 'C:/Malaria dataset/dataset/train'
TEST_DATA_DIR = 'C:/Malaria dataset/dataset/test'

# --- 3. Data Augmentation and Loading ---
# Data augmentation for TRAINING (to prevent overfitting)
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Only rescaling and normalization for TESTING/VALIDATION
test_datagen = ImageDataGenerator(rescale=1.0 / 255)

# Loading dataset for Training
train_generator = train_datagen.flow_from_directory(
    TRAIN_DATA_DIR,
    target_size=(IMG_WIDTH, IMG_HEIGHT),
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

# Loading dataset for Testing/Validation
test_generator = test_datagen.flow_from_directory(
    TEST_DATA_DIR,
    target_size=(IMG_WIDTH, IMG_HEIGHT),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False # CRITICAL: MUST be False for correct evaluation mapping
)

# Get class labels
class_labels = list(test_generator.class_indices.keys())
print(f"Detected Class Labels: {class_labels}")


# --- 4. CNN Model Definition and Compilation ---
model = Sequential([
    # First Convolutional Block
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_WIDTH, IMG_HEIGHT, 3)),
    MaxPool2D(pool_size=(2, 2)),
    Dropout(0.2),

    # Second Convolutional Block
    Conv2D(64, (3, 3), activation='relu'),
    MaxPool2D(pool_size=(2, 2)),
    Dropout(0.2),

    # Fully Connected Layers
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Binary classification output
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Display model architecture
model.summary()


# --- 5. Model Training and Saving ---
print(f"\n--- Starting Model Training for {EPOCHS} Epochs ---")
history = model.fit(
    train_generator, 
    epochs=EPOCHS, 
    validation_data=test_generator,
    verbose=1
)
print("--- Training Complete ---")

# Save the trained model
model.save(MODEL_FILENAME)
print(f"Model successfully saved as {MODEL_FILENAME}")


# --- 6. Plot Training History ---
# Plotting the model's accuracy over epochs
plt.figure(figsize=(10, 4))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()


# --- 7. Comprehensive Model Evaluation ---
print("\n--- Starting Comprehensive Model Evaluation ---")

# Generate Predictions
test_generator.reset() 
y_pred_probs = model.predict(test_generator, steps=len(test_generator), verbose=1)

# Convert probabilities to binary class predictions (0 or 1)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()

# Get the true labels
y_true = test_generator.classes 

# Classification Report (Precision, Recall, F1-Score)
print("\n--- Detailed Classification Report ---")
print(classification_report(y_true, y_pred, target_names=class_labels))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
cm_df = pd.DataFrame(cm, 
                     index = [f'True {label}' for label in class_labels], 
                     columns = [f'Pred {label}' for label in class_labels])

print("\n--- Confusion Matrix Data ---")
print(cm_df)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix for Malaria Classification')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

# AUC-ROC Curve
auc_score = roc_auc_score(y_true, y_pred_probs)
print(f"\nArea Under the ROC Curve (AUC): {auc_score:.4f}")

# Plot ROC curve
fpr, tpr, thresholds = roc_curve(y_true, y_pred_probs)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {auc_score:.4f})')
plt.plot([0, 1], [0, 1], 'r--', label='Random Guessing (AUC = 0.5)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate (Recall)')
plt.legend()
plt.show()


# --- 8. Single Image Prediction Function ---
def predict_image(image_path):
    """Loads a single image and makes a prediction using the trained model."""
    if not os.path.exists(image_path):
        print(f"Error: Image not found at {image_path}")
        return
        
    # Preprocess image
    img = image.load_img(image_path, target_size=(IMG_WIDTH, IMG_HEIGHT))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    
    # Predict
    prediction = model.predict(img_array)
    
    # Interpret result
    predicted_class_index = (prediction[0] > 0.5).astype(int)
    predicted_label = class_labels[predicted_class_index[0]]
    
    # Calculate confidence in the predicted class
    confidence = prediction[0][0] if predicted_class_index[0] == 1 else 1 - prediction[0][0]

    print(f"\nPrediction for {os.path.basename(image_path)}:")
    print(f"The cell is classified as: {predicted_label}")
    print(f"Confidence in this class: {confidence:.2f}")

# --- Example Prediction (Uncomment and set a valid path to test) ---
# Example_image_path = 'C:/Malaria dataset/dataset/test/Uninfected/C1_thinF_IMG_20150604_104925_cell_11.png'
# predict_image(Example_image_path)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk

# --- 1. Load the Model ---
MODEL_PATH = 'malaria_cnn_model.h5' # Ensure this path is correct
IMG_WIDTH, IMG_HEIGHT = 64, 64

try:
    loaded_model = tf.keras.models.load_model(MODEL_PATH)
    print("Model loaded successfully!")
except Exception as e:
    print(f"Error loading model: {e}")
    # Exit or disable prediction if model fails to load

def preprocess_image(img_path):
    """Loads, resizes, and normalizes the image for the model."""
    img = image.load_img(img_path, target_size=(IMG_WIDTH, IMG_HEIGHT))
    img_array = image.img_to_array(img) / 255.0  # Normalize
    img_array = np.expand_dims(img_array, axis=0) # Add batch dimension
    return img_array

def make_prediction(img_path):
    """Performs the prediction using the loaded model."""
    if 'loaded_model' not in globals():
        return "Error: Model not loaded.", 0.0
        
    processed_img = preprocess_image(img_path)
    prediction = loaded_model.predict(processed_img)
    
    # The model outputs a probability (0-1). > 0.5 is UNINFECTED (Class 1)
    confidence = prediction[0][0]
    
    if confidence > 0.5:
        result = "UNINFECTED"
        confidence_score = confidence
    else:
        result = "PARASITIZED"
        confidence_score = 1.0 - confidence # Show confidence in the chosen class (Infected)
        
    return result, confidence_score

# --- 2. Tkinter GUI Functions ---

def select_image():
    """Opens a file dialog to select an image."""
    file_path = filedialog.askopenfilename(
        title="Select Cell Image",
        filetypes=[("Image Files", "*.png *.jpg *.jpeg")]
    )
    if not file_path:
        return
        
    # Run Prediction
    result, confidence = make_prediction(file_path)
    
    # Update GUI Display
    display_image(file_path)
    result_label.config(
        text=f"Prediction: {result}\nConfidence: {confidence:.2f}",
        bg='red' if result == 'PARASITIZED' else 'green',
        fg='white'
    )

def display_image(path):
    """Displays the selected image in the GUI."""
    img = Image.open(path)
    # Resize image for display, e.g., to 200x200
    img = img.resize((200, 200), Image.Resampling.LANCZOS)
    img = ImageTk.PhotoImage(img)
    panel.config(image=img)
    panel.image = img # Keep a reference

# --- 3. GUI Setup ---

root = tk.Tk()
root.title("Malaria Cell Classifier")

# Components
select_button = tk.Button(root, text="Select Image for Diagnosis", command=select_image, font=('Arial', 14))
select_button.pack(pady=10)

# Panel to display the image
panel = tk.Label(root, text="Image Preview Here", width=30, height=10, relief="groove")
panel.pack(padx=10, pady=10)

# Label to display the result
result_label = tk.Label(root, text="Result: Awaiting Selection...", font=('Arial', 16, 'bold'))
result_label.pack(pady=20, padx=10, fill='x')

root.mainloop()
```
